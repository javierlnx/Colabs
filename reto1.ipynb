Esta matriz permite rotar un vector en ùë• ‚àà ‚Ñù2 en un √°ngulo ùúÉ expresado en radianes con respecto al
origen.

a) Muestre que ùëÖ(ùúÉ)‚àí1 = ùëÖ(ùúÉ)ùëá

b) Muestre que el determinante de ùëÖ(ùúÉ) es 1.

c) Implemente un c√≥digo en Python que reciba un vector de entrada ùë• ‚àà ‚Ñù2 y un √°ngulo en radianes ùúÉ, y devuelva el vector rotado. Adem√°s, este c√≥digo deber√≠a graficar el vector original
y el vector rotado (en colores diferentes).

```python
import numpy as np
import matplotlib.pyplot as plt

def rotate_vector(x, theta):
  """
  Rotates a 2D vector by an angle theta in radians.

  Args:
    x: A numpy array representing the 2D vector [x, y].
    theta: The angle in radians to rotate the vector.

  Returns:
    A numpy array representing the rotated vector.
  """
  rotation_matrix = np.array([[np.cos(theta), np.sin(theta)],
                              [-np.sin(theta), np.cos(theta)]])
  rotated_x = np.dot(rotation_matrix, x)
  return rotated_x

def plot_vectors(original_vector, rotated_vector):
  """
  Plots the original and rotated 2D vectors.

  Args:
    original_vector: A numpy array representing the original 2D vector.
    rotated_vector: A numpy array representing the rotated 2D vector.
  """
  plt.figure()
  plt.quiver(0, 0, original_vector[0], original_vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Original Vector')
  plt.quiver(0, 0, rotated_vector[0], rotated_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Rotated Vector')
  plt.xlim([-5, 5])
  plt.ylim([-5, 5])
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('Vector Rotation')
  plt.grid()
  plt.legend()
  plt.show()

# Example usage:
x = np.array([2, 1])
theta = np.pi/4 # Rotate by 45 degrees

rotated_x = rotate_vector(x, theta)
print("Original vector:", x)
print("Rotated vector:", rotated_x)

plot_vectors(x, rotated_x)
```
