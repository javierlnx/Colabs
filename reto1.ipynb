Considere la siguiente matriz:
ùëÖ(ùúÉ) = [ ùëêùëúùë†(ùúÉ) ùë†ùëñùëõ(ùúÉ)
        ‚àíùë†ùëñùëõ(ùúÉ) ùëêùëúùë†(ùúÉ)].
        
Esta matriz permite rotar un vector en ùë• ‚àà ‚Ñù2 en un √°ngulo ùúÉ expresado en radianes con respecto al
origen.

a) Muestre que ùëÖ(ùúÉ)‚àí1 = ùëÖ(ùúÉ)ùëá

b) Muestre que el determinante de ùëÖ(ùúÉ) es 1.

c) Implemente un c√≥digo en Python que reciba un vector de entrada ùë• ‚àà ‚Ñù2 y un √°ngulo en radianes ùúÉ, y devuelva el vector rotado. Adem√°s, este c√≥digo deber√≠a graficar el vector original
y el vector rotado (en colores diferentes).

```python
import numpy as np
import matplotlib.pyplot as plt

# Punto c) Rotaci√≥n de vector y graficaci√≥n ---

def rotate_vector(x, theta):
  #
  rotation_matrix = np.array([[np.cos(theta), np.sin(theta)],
                              [-np.sin(theta), np.cos(theta)]])
  rotated_x = np.dot(rotation_matrix, x)
  return rotated_x

def plot_vectors(original_vector, rotated_vector):
  """
  Grafica los vectores 2D original y rotado.

  Args:
    original_vector: Un array de numpy que representa el vector 2D original.
    rotated_vector: Un array de numpy que representa el vector 2D rotado.
  """
  plt.figure()
  plt.quiver(0, 0, original_vector[0], original_vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector Original')
  plt.quiver(0, 0, rotated_vector[0], rotated_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector Rotado')
  plt.xlim([-5, 5])
  plt.ylim([-5, 5])
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('Rotaci√≥n de Vector')
  plt.grid()
  plt.legend()
  plt.show()

# Ejemplo de uso:
x = np.array([2, 1])
theta = np.pi/4 # Rotar por 45 grados

rotated_x = rotate_vector(x, theta)
print("Vector original:", x)
print("Vector rotado:", rotated_x)

plot_vectors(x, rotated_x)

# --- Parte a) y b) Propiedades de la matriz de rotaci√≥n ---

# Parte a) Muestra que R(theta)^-1 = R(theta)^T
theta_example = np.pi / 3 # √Ångulo de ejemplo
R = np.array([[np.cos(theta_example), np.sin(theta_example)],
              [-np.sin(theta_example), np.cos(theta_example)]])

R_inv = np.linalg.inv(R)
R_T = R.T

print("\nR(theta)^-1:\n", R_inv)
print("R(theta)^T:\n", R_T)

# parte a) Verifica si R_inv es similar a R_T
print("\n¬øEs R(theta)^-1 similar a R(theta)^T?:", np.allclose(R_inv, R_T))

# Parte b) Muestra que el determinante de R(theta) es 1
det_R = np.linalg.det(R)
print("\nDeterminante de R(theta):", det_R)

# Verifica si el determinante es similar a 1
print("¬øEs el determinante de R(theta) similar a 1?:", np.isclose(det_R, 1.0))
```
