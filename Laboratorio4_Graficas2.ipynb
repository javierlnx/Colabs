## **2. Superficies y curvas de nivel**

Durante el curso se tratarn temas de optimizacin donde ser necesario el manejo de funciones multivariable. Muchos procesos optimizacin requieren un anlisis grficos para facilitar el entendimiento y verificar las soluciones. Entre estos anlisis, estn las superficies y curvas de nivel, las cuales son fundamentales para el curso. Para comenzar, es necesario comprender que estas funciones deben mapear desde $\mathbb{R^n}$ hasta $\mathbb{R}$. Es decir, deben tomar un vector de variables $[x_1, x_2, ..., x_n]^T$ y transformarlo en un escalar a travs de operaciones matemticas. Estas funciones de definen como:

\begin{align*}
f: \mathbb{R^n}  \rightarrow \mathbb{R}.
\end{align*}

Hace falta aclarar que para poder graficar una superficie, la funcin debe mapear desde $\mathbb{R^2}$ hasta $\mathbb{R}$, porque, de lo contrario, si $n>2$, no hay forma de graficar la funcin.

**2.1. Superficies**

 Para comenzar, se mostrar el proceso para graficar una superficie y los posibles ajustes y variantes que hay para esto. En principio, es necesario crear una funcin, en este caso, como ejemplo, se usar un paraboloide definido como:

\begin{align*}
f(x_1,x_2) = x_1^2 + x_2^2
\end{align*}

El proceso es similar a un grfico simple, pero con algunos parmetros adicionales. Primero, veremos cmo cambia la funcin *subplots*, a la cual se le debe adicional el parmetro:

```
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
```

Lo que indica que se trata de una superficie. Ahora, se definen las variables como arreglos, igual que en el grfico simple:



```
x_1 = np.linspace(-5,5,100)
x_2 = np.linspace(-5,5,100)
```

El principal cambio es que la funcin no se puede definir con los arreglos creados anteriormente, debido a que es necesario crear una cuadrcula de puntos que sern proyectados en un eje vertical. La forma de hacer esto es la siguiente:

```
(X_1, X_2) = np.meshgrid(x_1, x_2)
```

Ahora si es posible definir la funcin:

```
f = X_1**2 + X_2**2
```

Ahora, para graficar la superficie se usa la funcin *plot_surface*, la cual recibe como parmetros las variables independientes en forma de cuadrcula y la funcin dependiente.

```
surf = ax.plot_surface(X_1, X_2, f)
```

Sin embargo, esta puede recibir ms parmetros para poder ajustar el grfico. Algunos de ellos son:

- *rcount, ccount:* Indica la cantidad de muestras que toma de cada variable independiente. El valor por defecto es 50,
- *cmap:* Agrega efectos de color (Leer documentacin para ver las opciones)
- *facecolors:* Colorea un solo cuadrante de la cuadrcula.
- *shade:* Agrega sombra. (True por defecto).

Finalmente, es posible que el ttulo del eje vertical no sea visible. Para solucionar eso, se incluye el siguiente comando:

```
ax.set_box_aspect(aspect=None, zoom=0.8)
```
Y luego se ajusta el tamao de la figura.

El resultado del ejemplo sera el siguiente:


````python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(subplot_kw={"projection": "3d"} ,  figsize = (5, 5))
x_1 = np.linspace(-5,5,100)
x_2 = np.linspace(-5,5,100)
(X_1, X_2) = np.meshgrid(x_1, x_2)
f = X_1**2 + X_2**2

surf = ax.plot_surface(X_1, X_2, f)
ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_zlabel("Z label")
ax.set_box_aspect(aspect=None, zoom=0.8)
ax.set(title = 'Paraboloide')

plt.show()
````

Igual que en los grficos simples, existen variantes para graficar una superficie. La forma de incluirlas, es cambiando la funcin *plot_surface* por alguna de estas:

- *scatter:* Grafica los datos como puntos.
- *bar3d:* Grafica los datos en barras.
- *wireframe:* Cambia la superficie por un alambrado.
- *trisufr*: Modifica la cuadrcula por tringulos.
- *errorbar*: Aade barras de error en los puntos.


**2.2. Curvas de nivel**

En ocasiones, en anllisis grfico no es suficiente a travs de la representacin de la funcin como una superficie. En vez de esto, es resulta ms conveniente representarla mediante curvas de nivel. Las curvas de nivel son el resultado de proyectar en un plano bidimensional un corte vertical del slido generado por funcin, como se muestra en la siguiente imgen:

image.png

Note que, en el caso de un paraboloide,  el corte transversal a lo largo del eje vertical genera una circunferencia. En esencia, hay dos formas de representar estas curvas: a travs de la curva generada, o mediante un mapa de calor. La forma de hacerlo es mediante las funciones *contour* y *contourf*, respectivamente, y su sintaxis es:

```
CS = ax.contour(X_1, X_2, f)
CS = ax.contourf(X_1, X_2, f)
```

Igual que todas las funciones anteriores, estas tambin se pueden ajustar mediante parmetros. Algunos de ellos son:

Para ambas funciones:

- *levels*: Indica el nmero de cortonos que se grafica.
- *alpha*: Indica la transparencia de los colores. (*aplpha* = 0: invisible, *alpha = 1: visible)
- *colors*: Indica el color de las curvas.
- *cmap*: Cambia el tipo de efecto de color.

Adicionalmente, cuando se trabaja con curvas de nivel, es posible usar otras funciones tiles como *clabel*, la cual permite agregar ttulos a las curvas similar a una leyenda. La forma de usarla es la siguiente:

```
ax.clabel(CS)
```

Donde algunos argumentos de la funcin son:

- *fontsize*: Ajusta el tamao de la letra.
- *colors*: Especifica el color de la letra.
- *rightside_up*: Cambia la orientacin del texto de forma radial (True por defecto)

Finalmente, para el caso de la funcin *contourf*, es conveniente agregar una barra de color que indique cuantitativamente el rango de los colores. La forma de hacerlo es:

```
fig.colorbar(CS)
```

Usando el ejemplo del paraboloide, quedara as:


````python
fig, ax = plt.subplots()
x_1 = np.linspace(-5,5,100)
x_2 = np.linspace(-5,5,100)
(X_1, X_2) = np.meshgrid(x_1, x_2)
f = X_1**2 + X_2**2

CS1 = ax.contour(X_1, X_2, f, alpha = 0.5)
ax.clabel(CS1, rightside_up = False)
ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set(title = 'Contour')
plt.show()


fig, ax = plt.subplots()
CS2 = ax.contourf(X_1, X_2, f, levels = 10)
ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set(title = 'Contourf')
fig.colorbar(CS2)
plt.show()
````

**2.3.1.** Grafique una superficie para las siguientes funciones cuadrticas en cuatro subplots en una misma figura (recuerde aadir labels para los tres ejes, ttulos a los subplots y ttulo general):

- $f_1(x_1, x_2) = x_1^2 - x_2^2$, (surface, agregue coloreado tipo *coolwarm*)
- $f_2(x_1, x_2) = - x_1^2 + x_2^2$, (scatter)
- $f_3(x_1, x_2) = x_1^2 + x_2^2 + x_1x_2$, (wireframe)
- $f_4(x_1, x_2) = x_1^2 + x_2^2 - x_1x_2$, (trisurf)

Donde $x_1 \in [-10,10], x_2 \in [-10,10]$

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for x1 and x2
x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)

# Create a meshgrid
X1, X2 = np.meshgrid(x1, x2)

# Define the functions
f1 = X1**2 - X2**2
f2 = -X1**2 + X2**2
f3 = X1**2 + X2**2 + X1 * X2
f4 = X1**2 + X2**2 - X1 * X2

# Create a figure with 4 subplots
fig, axes = plt.subplots(2, 2, subplot_kw={"projection": "3d"}, figsize=(10, 10))
fig.suptitle('Quadratic Surfaces')

# Plot f1 as a surface with coolwarm colormap
ax1 = axes[0, 0]
ax1.plot_surface(X1, X2, f1, cmap='coolwarm')
ax1.set_xlabel("X1")
ax1.set_ylabel("X2")
ax1.set_zlabel("f1")
ax1.set_title("f1(x1, x2) = x1^2 - x2^2")

# Plot f2 as scatter points
ax2 = axes[0, 1]
ax2.scatter(X1, X2, f2)
ax2.set_xlabel("X1")
ax2.set_ylabel("X2")
ax2.set_zlabel("f2")
ax2.set_title("f2(x1, x2) = -x1^2 + x2^2")

# Plot f3 as a wireframe
ax3 = axes[1, 0]
ax3.plot_wireframe(X1, X2, f3)
ax3.set_xlabel("X1")
ax3.set_ylabel("X2")
ax3.set_zlabel("f3")
ax3.set_title("f3(x1, x2) = x1^2 + x2^2 + x1*x2")

# Plot f4 as a trisurf
ax4 = axes[1, 1]
ax4.plot_trisurf(X1.flatten(), X2.flatten(), f4.flatten())
ax4.set_xlabel("X1")
ax4.set_ylabel("X2")
ax4.set_zlabel("f4")
ax4.set_title("f4(x1, x2) = x1^2 + x2^2 - x1*x2")

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
plt.show()
````
