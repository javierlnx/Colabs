## **1. Grficos simples**

Para comenzar la introduccin a la librera *Matplotlib*, deben graficar curvas simples y conjuntos de datos en un solo grfico con los respectivos nombres de los ejes, ttulo, leyendas y escalas. En general, para este punto, deben importar *pyplot*, de la siguiente forma:



```
import matplotlib.pyplot as plt
```

**1.1. Para ejemplificar la forma de realizar grficos simples, se grafican las funciones $sin(t)$ y $cos(t)$, donde $t$ es el tiempo. Para esto, se crea primero un arreglo con que contenga los datos de tiempo**

````python
import numpy as np

t = np.linspace(0,10,100)
````

````python
import matplotlib.pyplot as plt
````

````python
##METODO 1
plt.figure()
plt.plot(t,np.sin(t),"r",label="Seno")
plt.plot(t,np.cos(t),label="Coseno")
plt.grid()
plt.xlabel("t(seg)")
plt.ylabel("f(t)")
plt.title("Metodo 1")
plt.legend()
plt.show()
````

````python

##METODO 2

fig, ax = plt.subplots()
ax.plot(t,np.sin(t),label="Seno")
ax.set_title("Seno")

ax2 = ax.twinx()
ax2.plot(t,np.cos(t),label="Coseno")
"""ax.set_title("Seno")
ax.set_xlabel("t(seg)")"""
fig.suptitle("Funciones trigonometricas")

"""ax.grid()
ax.set_xlabel("t(seg)")
ax.set_ylabel("f(t)")
ax.set_title("Metodo 1")
ax.legend()
ax.grid()"""
fig.set_tight_layout(True)
plt.show()
````

Luego, se usa la librera *Numpy* para obtener los datos en la imgen de las funciones mencionadas

````python
s = np.sin(t)
c = np.cos(t)
````

Ahora, se debe conocer la funcin *subplots()*, la cual permite retornar los atributos *fig* y *ax* con diversos parmetros para poder configurar el grfico. Algunos parmetros de esta funcin son:

- *nrows*, *ncols*: Nmero de filas y columnas en una cuadrcula de subplots (Valor 1 por defecto)
- *sharex, sharey*: Permite compartir eje *x* y eje *y* (Valor por defecto 'False')
- *figsisze*: Permite establecer un tamao del grfico. *figsize = (Height, width)*
- *layout* = Ajusta la cuadrcula para evitar que se sobrelapen los sublots. *layout = 'constrained'*

La forma de usar esta funcin es:

```
fig, ax = plt.subplots(nrows, ncols, sharex = False/True, sharey = False/True, figsize = (height, width))
```


Debido a que solo hay un subplot, el atributo *ax* solo tiene un elemento que corresponde a dicho subplot. Luego, para graficar lo anterior, a este atributo, es posible aplicarle algunas funciones para graficar. Algunas de ellas son:


- *ax.plot(x,y,label = 'leyenda para la curva', linewidth = Grosor de curva)* : Grafica el conjunto de datos *y* contra el conjunto *x*
- *ax.set(xlabel = 'Ttulo del eje x', ylabel = 'Ttulo del eje y', title = 'Ttulo del subplot')*: Asigna ttulos a los ejes y al subplot
- *ax.legend(loc = 'Ubicacin de la leyenda')*: Muestra la leyenda de las curvas graficadas, la ubicaciones pueden ser: *'right', 'left', 'center', 'upper right', 'upper left', 'lower right', 'lower left', 'center right', 'center left', 'upper center, 'lower center'*
- *ax.grid()*: Muestra una cuadrcula.
- *ax.set_xlim([mnimo, mximo])*: Establece lmites para el eje $x$.
- *ax.set_ylim([mnimo, mximo])*: Establece lmites para el eje $y$.

Cabe resaltar que hay otras formas de configurar el grfico para llegar al mismo resultado, depende del estilo de cada programador.

Luego, para darle un ttulo general al grfico, se hace de la forma:

```
fig.suptitle('Ttulo general del grfico')
```
Finalmente, para mostrar el grfico, se escribe:

```
plt.show()
```
A continuacin, se muestra el ejemplo anterior:


````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(t,s,label='Funcin seno')
ax.plot(t,c,label='Funcin coseno')
ax.set(xlabel = 'Tiempo [s]', ylabel='Funciones', title = 'Seno vs Coseno')
ax.legend(loc='lower right')
fig.suptitle('Grfico simple')
plt.show()
````

Si se quisera presentar las curvas en dos subplots diferentes, hay que tener en cuenta que al cambiar el nmero de filas y/o columnas, el atributo *ax* cambia su tamao para ser un arreglo. Por ejemplo, si se quisiera presentar un grfico de dos filas y dos columnas, la forma sera:

```
fig, ax = plt.subplots(2, 2)
```

Donde *ax* ahora es un arreglo de tamao $2 \times 2$, luego, si se quiere llamar algn subplot en especfico, por ejemplo, el que se encuentra en la posicin $(2,1)$, se hace de la forma:

```
ax[1,0]
```

A continuacin, se muestra la forma en la que se presentaran las curvas anteriroes en dos subplots:

````python
fig, ax = plt.subplots(1 , 2)

ax[0].plot(t, s)
ax[0].set_title('Funcin Seno')
ax[0].set(xlabel = 'Tiempo [s]', ylabel='Funciones')


ax[1].plot(t, c)
ax[1].set_title('Funcin Coseno')
ax[1].set(xlabel = 'Tiempo [s]', ylabel='Funciones')
ax[1].label_outer()

fig.suptitle('Grfico simple')
plt.show()
````

Es necesario aclarar que hay diferentes formas de graficar un conjunto de datos. Para esto, se remplaza la funcin *plot* por otras como:

- *scatter*: Grafica los conjuntos de datos como puntos.
- *bar*: Grafica los conjuntos como barras.
- *errorbar*: Incluye barras de error para datos estadsticos.
- *fill_between*: Sobrea reas de confianza.
- *hist*: Muestra un histograma de los datos.

La documentacin de estas funciones se encuentra en el siguiente link: [Documentacin](https://matplotlib.org/stable/gallery/index.html)

**1.2.1.** Grafique las siguientes funciones en cuatro subplots, incluya labels, ttulos en subplots y ttulo para la figura:

- $f(x) = \ln(x) $, (Ttulo: Logartmica)
- $g(x) = e^x$, (Ttulo: Exponencial)
- $f(x) + g(x)$, (Ttulo: Suma)
- $f(g(x))$, (Ttulo: Compuesta)

Donde $x$ est en segundos y $f(x), g(x)$ estn en metros.



````python
fig, ax = plt.subplots(1, 4)

#()=ln() , (Ttulo: Logartmica), Donde    est en segundos y  (),()  estn en metros.
ax[0].plot(t[1:], np.log(t[1:]))
ax[0].set_title('Logartmica')
ax[0].set(xlabel = 'Tiempo [s]', ylabel='Funciones')

#()= , (Ttulo: Exponencial)
ax[1].plot(t, np.exp(t))
ax[1].set_title('Exponencial')
ax[1].set(xlabel = 'Tiempo [s]', ylabel='Funciones')

# ()+() , (Ttulo: Suma)
ax[2].plot(t[1:], np.log(t[1:]) + np.exp(t[1:]))
ax[2].set_title('Suma')
ax[2].set(xlabel = 'Tiempo [s]', ylabel='Funciones')

# (()) , (Ttulo: Compuesta)
ax[3].plot(t, np.log(np.exp(t)))
ax[3].set_title('Compuesta')
ax[3].set(xlabel = 'Tiempo [s]', ylabel='Funciones')

fig.suptitle('Grfico simple')
plt.show()
````

**1.2.3.** Suponga que hizo una regresin lineal de la posicin en el tiempo de un atelta, donde obtuvo la siguiente ecuacin:

\begin{align*}
  y = 3x - 1
\end{align*}

Tambin sabe que el error de todos los datos es de $0.3$. Grafique 10 puntos de forma *scatter* donde incluya barras de error en cada punto. Adems, el eje horizontal debe estar entre $5$ y $10$.

````python
# Generate 10 random x values between 5 and 10
x = 5 + 5 * np.random.rand(10)

# Calculate the corresponding y values based on the equation y = 3x - 1
y = 3 * x - 1

# Define the error for all data points
error = 0.3

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the data points with error bars as a scatter plot
ax.errorbar(x, y, yerr=error, fmt='o', label='Data with Error')

# Set the x-axis limits
ax.set_xlim([5, 10])

# Add labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Linear Regression with Error Bars')
ax.legend()
ax.grid(True)

# Show the plot
plt.show()
````

**1.2.4.** Suponga que la regresin anterior no fue suficiente y necesita tambin graficar la regresin de la frecuencia cardiaca, que sigue la siguiente relacin:

\begin{align*}
  y = 70 + 2\sqrt{10x}
\end{align*}

Grafique sobre el mismo subplot pero agregue un eje vertical a la derecha correspondiente a la frecuencia cardiaca en uniades de $\frac{beat}{min}$. Incluya leyendas. El eje asociado a la posicin debe ir desde 70 hasta 115.

````python
# Assuming x and y from the previous plot are still available
# If not, you would need to regenerate them here:
# x = 5 + 5 * np.random.rand(10)
# y = 3 * x - 1
# error = 0.3

fig, ax1 = plt.subplots()

# Plot the data points with error bars on the first y-axis
ax1.errorbar(x, y, yerr=error, fmt='o', label='Posicin con Error')
ax1.set_xlabel('Tiempo [s]')
ax1.set_ylabel('Posicin [m]')
ax1.set_ylim([70, 115]) # Set the y-axis limits for the position

# Create a second y-axis
ax2 = ax1.twinx()

# Calculate the frequency response
y_freq = 70 + 2 * np.sqrt(10 * x)

# Plot the frequency response on the second y-axis
ax2.plot(x, y_freq, 'r-', label='Frecuencia Cardiaca')
ax2.set_ylabel('Frecuencia Cardiaca [beat/min]')

# Add title and legends
plt.title('Regresin Lineal y Frecuencia Cardiaca')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.grid(True)
plt.show()
````

*1.2.5* Cree 1000 datos aleatorios con distribucin uniforme y otros 1000 con distrbucin normal y smelos de la siguiente forma:



```
uniform_Set = np.random.uniform(size = 1000)
normal_Set = np.random.normal(size = 1000)
sum = np.add(Uniform_Set, Normal_Set)

```

Grafique el correspondiente histograma para 10, 20, 30 y 50 particiones (bins), cada una en un subplot diferente. Agregue ttulo al grfico, a los subplots, y a los ejes (invente el ttulo para estos).

````python
# Create 1000 random data points with uniform and normal distributions
uniform_Set = np.random.uniform(size=1000)
normal_Set = np.random.normal(size=1000)

# Sum the two sets of data
sum_Set = np.add(uniform_Set, normal_Set)

# Create a figure with four subplots
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

# Define the number of bins for each subplot
bins = [10, 20, 30, 50]

# Plot histograms in each subplot
for i, n_bins in enumerate(bins):
    axes[i].hist(sum_Set, bins=n_bins)
    axes[i].set_title(f'Histograma con {n_bins} bins')
    axes[i].set_xlabel('Valor')
    axes[i].set_ylabel('Frecuencia')

# Add a general title to the figure
fig.suptitle('Histogramas de la suma de distribuciones Uniforme y Normal')

# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()

# Show the plot
plt.show()
````
